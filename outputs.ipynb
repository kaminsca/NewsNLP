```python
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, f1_score
import pandas as pd
import seaborn as sns
import matplotlib as plt

df = pd.read_csv("./data/test.csv", delimiter="|")
columns = ["avg_white_pop_pct","avg_median_hh_inc","avg_non_college_pct"]
labels = np.array(df[columns].values.tolist())
preds = []
for i in range(35000):
    preds.append([])
    for j in range(3):
        preds[i].append(1 if np.random.random() >= 0.5 else 0)
preds = np.array(preds)
acc = [accuracy_score(labels.flatten()[i * 35000:(i+1) * 35000], preds.flatten()[i * 35000:(i+1) * 35000]) for i in range(3)]
precs = precision_score(labels, preds, average=None)
f1s = f1_score(labels, preds, average=None)
f1m = f1_score(labels, preds, average='micro')

# label1 label2 label3
# acc, prec, f1, f1m
# random
a1 = [0.501, 0.502, 0.500]
a2 = [0.528, 0.564, 0.495]
a3 = [0.514, 0.528, 0.499]
a4 = 0.514

# svm
b1 = [0.740, 0.753, 0.707]
b2 = [0.704, 0.756, 0.682]
b3 = [0.734, 0.693, 0.729]
b4 = 0.733

# knn
c1 = [0.717, 0.710, 0.697]
c2 = [0.722, 0.730, 0.679]
c3 = [0.744, 0.744, 0.704]
c4 = 0.732

# bert
d1 = [0.809, 0.767, 0.842]
d2 = [0.758, 0.934, 0.839]
d3 = [0.905, 0.774, 0.83]
d4 = 0.862

# electra
e1 = [0.968, 0.796, 0.774]
e2 = [0.913, 0.957, 0.766]
e3 = [0.943, 0.803, 0.914]
e4 = 0.921

df_data = []
for i in range(3):
    df_data.append(['random', 'acc', sum(a1) / len(a1)])
    df_data.append(['random', 'prec', sum(a2) / len(a2)])
    df_data.append(['random', 'f1m', a4])

    df_data.append(['svm', 'acc', sum(b1) / len(b1)])
    df_data.append(['svm', 'prec', sum(b2) / len(b2)])
    df_data.append(['svm', 'f1m', b4])

    df_data.append(['knn', 'acc', sum(c1) / len(c1)])
    df_data.append(['knn', 'prec', sum(c2) / len(c2)])
    df_data.append(['knn', 'f1m', c4])

    df_data.append(['bert', 'acc', sum(d1) / len(d1)])
    df_data.append(['bert', 'prec', sum(d2) / len(d2)])
    df_data.append(['bert', 'f1m', d4])

    df_data.append(['electra', 'acc', sum(e1) / len(e1)])
    df_data.append(['electra', 'prec', sum(e2) / len(e2)])
    df_data.append(['electra', 'f1m', e4])

df = pd.DataFrame(data=df_data, columns=['model', 'metric', 'value'])
df.head()
```
